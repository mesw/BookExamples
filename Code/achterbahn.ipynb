```python
def achterbahn(start):
    running = start
    bahn = [start]
    maximum = start
    iterations = 0
    while running > 1:
        iterations += 1
        if running % 2 == 0:
            running = running / 2
        else:
            running = running * 3 + 1
        bahn.append(running)
        if running > maximum:
            maximum = running
    return (bahn, maximum, iterations)
```

```python
b,m,i = achterbahn(4096)
```

```python
def maxiterbahn(start, stop):
    max_i = 0
    max_m = 0
    start_i = start
    start_m = start
    bahn_i = []
    bahn_m = []
    for s in range(start, stop):
        b,m,i = achterbahn(s)
        if m > max_m:
            max_m = m
            start_m = s
            bahn_m = b
        if i > max_i:
            max_i = i
            start_i = s
            bahn_i = b
    return max_i, start_i, bahn_i, max_m, start_m, bahn_m 
```

```python
max_i, start_i, bahn_i, max_m, start_m, bahn_m = maxiterbahn(1,32000)
```
